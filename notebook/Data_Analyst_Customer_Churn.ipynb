Charge Amount: Ordinal attribute (0: lowest amount, 9: highest amount)

# Train Data

```python
import os
import datetime
import pickle
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import AdaBoostClassifier, GradientBoostingClassifier
from xgboost import XGBClassifier
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score, recall_score, f1_score
```

```python
split = 30

class train_test_data():

  model = {"SVM":SVC(),
           "Logistic Regression":LogisticRegression(solver='liblinear'),
           "AdaBoost":AdaBoostClassifier(),
           "Gradient Boosting":GradientBoostingClassifier(),
           "Xgboost":XGBClassifier(objective= 'binary:logistic', nthread=4, seed=42)}

  def __init__(self, X_train, X_test, y_train, y_test):
    self.X_train = X_train
    self.X_test = X_test
    self.y_train = pd.Series(data=y_train)
    self.y_test = y_test

  def split_train(self, classifier):
    sss = StratifiedKFold(n_splits=split, random_state=None, shuffle=False)
    acc = []
    recall = []
    f1 = []
    pred_acc = []
    pred_recall = []
    pred_f1 = []
    result = {'accuracy':[],'recall':[], 'f1-score':[]}
    predict = {'pred_accuracy':[], 'pred_recall':[], 'pred_f1':[], 'pred_class':[]}
    for train_index, test_index in sss.split(self.X_train, self.y_train):
      xtrain, xtest = self.X_train.iloc[train_index], self.X_train.iloc[test_index]
      ytrain, ytest = self.y_train.iloc[train_index], self.y_train.iloc[test_index]
      clf_ = classifier.fit(xtrain, ytrain)
      pred = clf_.predict(xtest)
      acc.append(accuracy_score(pred, ytest))
      recall.append(recall_score(pred, ytest))
      f1.append(f1_score(pred, ytest))
      result['accuracy'] = acc
      result['recall'] = recall
      result['f1-score'] = f1

    tmp = clf_.predict(self.X_test)
    pred_acc.append(accuracy_score(self.y_test, tmp))
    pred_recall.append(recall_score(self.y_test, tmp))
    pred_f1.append(f1_score(self.y_test, tmp))
    predict['pred_accuracy'] = pred_acc
    predict['pred_recall'] = pred_recall
    predict['pred_f1'] = pred_f1
    predict['pred_class'] = tmp
    return result, predict

  def train_test_run(self):
    scores_res = []
    predict_res = []
    for classifier in list(train_test_data.model.values()):
      result, predict = self.split_train(classifier)
      scores_res.append(result)
      predict_res.append(predict)
    return scores_res, predict_res

  def get_score(self):
    scores_res, predict_res = self.train_test_run()

    #train data
    acc = list(zip(scores_res[0]['accuracy'], scores_res[1]['accuracy'], scores_res[2]['accuracy'], scores_res[3]['accuracy'], scores_res[4]['accuracy']))
    f1 = list(zip(scores_res[0]['f1-score'], scores_res[1]['f1-score'], scores_res[2]['f1-score'], scores_res[3]['f1-score'], scores_res[4]['f1-score']))
    df_acc = pd.DataFrame(acc, columns= list(train_test_data.model.keys()))
    df_f1 = pd.DataFrame(f1, columns= list(train_test_data.model.keys()))
    acc_score = df_acc.describe().T.sort_values(by=['mean'], ascending=False)
    f1_score = df_f1.describe().T.sort_values(by=['mean'], ascending=False)

    #predict data
    predict_acc = [predict_res[0]['pred_accuracy'][0], predict_res[1]['pred_accuracy'][0], predict_res[2]['pred_accuracy'][0], predict_res[3]['pred_accuracy'][0] ,predict_res[4]['pred_accuracy'][0]]
    predict_recall = [predict_res[0]['pred_recall'][0], predict_res[1]['pred_recall'][0], predict_res[2]['pred_recall'][0], predict_res[3]['pred_recall'][0], predict_res[4]['pred_recall'][0]]
    predict_f1 =  [predict_res[0]['pred_f1'][0], predict_res[1]['pred_f1'][0], predict_res[2]['pred_f1'][0], predict_res[3]['pred_f1'][0], predict_res[4]['pred_f1'][0]]
    predict_class = [predict_res[0]['pred_class'], predict_res[1]['pred_class'], predict_res[2]['pred_class'], predict_res[3]['pred_class'], predict_res[4]['pred_class']]
    pred_result_model = {'test accuracy':predict_acc, 'recall': predict_recall, 'f1-score':predict_f1}
    df_pred_res = pd.DataFrame(pred_result_model, index= list(train_test_data.model.keys()))
    df_pred_res.index.name = 'Model'
    sort = df_pred_res.sort_values(by=['f1-score'],ascending=[False])

    return [df_acc, acc_score, f1_score, sort]

```

```python
def save_df(df, save_dir):
  cr_date = datetime.datetime.now().strftime("%Y%m%d")
  df.to_csv(f'{save_dir}_{cr_date}.csv', index=False)

def get_model(df_predict):
  cr_date = datetime.datetime.now().strftime("%Y%m%d")
  model_name = df_predict.index[0]
  model = train_test_data.model[model_name]
  save_model_name = model_name + cr_date +'.pkl'
  pickle.dump(model, open(save_model_name, 'wb'))
```
